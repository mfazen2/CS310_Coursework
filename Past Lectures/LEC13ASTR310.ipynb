```python
x = np.linspace(-np.pi,np.pi,10)
y = np.linspace(-np.pi,np.pi,10)

x, y= np.meshgrid(x,y)
def func(x, y):
    return np.sin(x)*np.cos(y)
z = func(x, y)

s = z[3:7,3:7].sum()
print("sum:",s)

def f(x):
    return np.square(x)
squared = f(z)
print(squared)
print(f(2))

maxyval = np.max(z,0)
print(maxyval)
```

```python
import numpy.random as npr
x = y = npr.rand(100)
xx, yy = np.meshgrid(x,y)
print(x, y)
z = np.exp(-(xx**2+yy**2)/0.05)
print(z)
```

```python
hist,bin_edges =np.histogram(z,bins=10)
numcount = 1
edgecount = 0.0
print(bin_edges)
for val in hist:
    print(f"{numcount:2}). {edgecount:2.1f} {edgecount+.1:2.1f}{val:7}")
    numcount+=1
    edgecount+=.1
```
